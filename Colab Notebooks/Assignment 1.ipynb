10. Read an image and trim it, using Array Slicing.



```python
#Q10
from PIL import Image
import matplotlib.pyplot as plt
image=Image.open('x1.jpg')

image_slice=np.asarray(image.copy())[50:,:250,:]

fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(30,10))
ax[0].imshow(image)
ax[0].axis('off')
ax[1].imshow(image_slice)
ax[1].axis('off')
```

11. Paste a slice of an image onto another image.



```python
#Q11
image2=Image.open('x2.jpg')
image2.paste(image,(0,0))
plt.imshow(image2)
```

12. Flip an image.

```python
#Q12
fig,ax = plt.subplots(nrows=3,ncols=1,figsize=(30,10))
ax[0].imshow(np.flipud(image))
ax[0].axis('off')
ax[1].imshow(np.fliplr(image))
ax[1].axis('off')
ax[2].imshow(np.flip(image))
ax[2].axis('off')
```
