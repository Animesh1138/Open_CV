7. Implement a program to create a clip line using the cv2.clipLine() function.


```python
img_blank = np.zeros(shape=(512,512,3),dtype = np.int16)
ret, p1, p2 = cv2.clipLine((0, 0, 200, 200), (0, 0), (400, 400))
if ret:
    cv2.line(img = img_blank,pt1= p1, pt2 = p2, color=(0,255,0), thickness = 3)
plt.imshow(img_blank)
```
