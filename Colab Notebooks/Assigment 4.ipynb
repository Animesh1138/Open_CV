```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

#cv2.imshow works in Jupyter Lab but is disabled in Colab


cam = cv2.VideoCapture(0)

height = int(cam.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cam.get(cv2.CAP_PROP_FRAME_WIDTH))

writer = cv2.VideoWriter('/content/Cam_connected.mp4', cv2.VideoWriter_fourcc(*'XVID'),20, (width, height))


while True:
    res,frame = cam.read()
    writer.write(frame)
    cv2.imshow("frame",frame)
    if( cv2.waitKey(10) & 0xFF == 27):
        break



cam.release()
writer.release()
cv2.destroyAllWindows()
```
