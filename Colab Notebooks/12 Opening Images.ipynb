```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
img = cv2.imread('/content/00-puppy.jpg')
img
```

```python
plt.imshow(img)
#this happend as cv2 takes old convention of Imgae i.e. BGR instead of RGB
```

```python
cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('/content/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
#dark values as BLUE and lighter values as Green
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img_sqz=cv2.resize(img_gray,(1500,400))
```

```python
plt.imshow(img_sqz,cmap='gray')
```

```python
w_ratio = 0.3
h_ratio = 0.3 #of orignal height
img_resize = cv2.resize(img_rgb,(0,0),img_rgb,w_ratio,h_ratio) 
plt.imshow(img_resize)
```

```python
img_flip = cv2.flip(img_rgb,0) #0 for ud 1 for lr
plt.imshow(img_flip)
```
