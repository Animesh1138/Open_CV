```python
import cv2
sift = cv2.xfeatures2d.SIFT_create()
```

```python
obj = cv2.imread('/content/reeses_puffs.png',0)
Img = cv2.imread('/content/many_cereals.jpg',0)
sift = cv2.SIFT_create()

kp_obj,ds_obj = sift.detectAndCompute(obj,None)
kp_img,ds_img = sift.detectAndCompute(Img,None)

# Brute force
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)

matches = bf.knnMatch(ds_obj,ds_img,k=2)

# matches = sorted(matches,key = lambda  x:x.distance)

Img_sift = cv2.drawMatchesknn(obj,kp_obj,Img,kp_img,matches[:25],None,-1)

plt.imshow(Img_sift)
```
