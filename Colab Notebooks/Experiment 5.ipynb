# *Theory*

When a signal is processed, they signal undergoes many manipulations involving the independent variable and the dependent variable. 

Some of these manipulations include:
1.  Shifting the signal in the time domain.
2.  Folding the signal
3.  Scaling in the time-domain

A brief introduction of these manipulations is given below.


# Shifting
In the case of discrete-time signals, the independent variable is the time, t. A signal x(t) may be shifted in time, i.e. the signal can be either advanced in the time axis or delayed in the time axis. The shifted signal is represented by x(t-to), where k is an integer. 

If 'to' is positive, the signal is delayed by to units of time and if 'to' is negative, the time shift results in an advance of signal by 'to' units of time.


# Folding
This operation is done by replacing the independent variable t by –t. This results in folding of the signal about the origin, i.e. t=0. Folding is also known as the reflection of the signal about the time origin t=0. Folding of a signal is done while convoluting the signal with another.


# Time scaling
This involves replacing the independent variable t by at, where a is an integer. This process is also called as down sampling. If x(t) is the signal , then x(at) is called time scaled transformation of x(t). 

If a > 1, it implies that the signal is compressed and a < 1 implies the signal is expanded.

## Let's start with generating few signals first.

1. Generate a signal x1(t) = cos(6πt) and plot for 0 ≤ t ≤ 1







```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0,1,.01)
x1 = np.cos(6*(np.pi)*t)
plt.figure(figsize=(8, 2))
plt.plot(t,x1, label ='cos(6πt)');plt.ylabel('x1(t)=cos(6πt)')
plt.title('Cosine curve');plt.legend();plt.xlabel('t');plt.show()
```

2. Plot four periods of signal x2(t) = 3cos(3πt+π/3) 

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0,3,.01)
x2 = 3*(np.cos(3*(np.pi)*t + (np.pi)/3));plt.figure(figsize=(8, 2))
plt.plot(t,x2);plt.ylabel('x2(t)= 3cos(3πt+π/3)')
plt.title('Cosine curve');plt.xlabel('t');plt.show()
```


6. Find even and odd components of signal x3(t) as defined above in Problem 2.


```python
import matplotlib.pyplot as plt
import numpy as np
def x3(t):
  return 3*np.exp(0.4*t)
t = np.arange(-2,2,.2)
plt.plot(t,x3(t), label ='x3(t)')
x3_even = .5*(x3(t) + x3(-t));x3_odd = .5*(x3(t) -x3(-t))
plt.plot(t,x3_even, label ='Even Signal of x3(t)');
plt.plot(t,x3_odd, label ='Odd Signal of x3(t)')
plt.plot(t,x3_even+x3_odd, 'r*', label ='reconstructed x3(t)')
plt.ylabel('Even and Odd decomposition of a signal and its reconstruction')
plt.xlabel('t');plt.legend();
plt.show()
```
